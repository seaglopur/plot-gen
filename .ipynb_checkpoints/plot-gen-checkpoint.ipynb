```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
fig = plt.gcf()
fig.set_size_inches(12.5, 7.5)

N = 3
Atingidos = (95, 94, 89)
Violados = (5, 6, 11)
ind = np.arange(N)    
width = 0.70       

p1 = plt.bar(ind, Atingidos, width, color='#90EE90', zorder=3)
p2 = plt.bar(ind, Violados, width, color='#CD5C5C',
             bottom=Atingidos, zorder=3)

plt.xlabel('Mês/Ano (Ticket Resolvido)')
plt.grid(b=None, which='major', linestyle='-.', axis='y')
plt.xticks(ind, ('Oct 2018', 'Nov 2018', 'Dez 2018'))
plt.yticks(np.arange(0, 101, 25))
plt.legend((p1[0], p2[0]), ('% Atingidos', '% Violados'))
plt.gca().yaxis.set_major_formatter(PercentFormatter())
for p in plt.gca().patches:
    if p.get_height() > 50:
        plt.gca().annotate(str(p.get_height())+'%', (p.get_x() + 0.90*p.get_width()/2, p.get_y() + p.get_height()*2/3))

plt.show()
```
