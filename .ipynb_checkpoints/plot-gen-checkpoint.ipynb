```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
fig = plt.gcf()
fig.set_size_inches(12.5, 7.5)

# Número de blocos no gráfico
N = 3

# Porcentagem de SLA atingido, respectivamente, nos blocos: (porcentagembloco1, porcentagembloco2, porcentagembloco3)
Atingidos = (100, 100, 89)

# Porcentagem de SLA violado, respectivamente, nos blocos: (porcentagembloco1, porcentagembloco2, porcentagembloco3)
# Caso a porcentagem de SLA violado seja "0", colocar como valor "0.1", para melhor diagramação do gráfico.
Violados = (0.1, 0.1, 11)

ind = np.arange(N)    
width = 0.70       

p1 = plt.bar(ind, Atingidos, width, color='#99cc99', zorder=3)
p2 = plt.bar(ind, Violados, width, color='#ff7f7f',
             bottom=Atingidos, zorder=3)

plt.xlabel('Mês/Ano (Ticket Resolvido)', color='#292929', fontsize=18)
plt.grid(b=None, which='major', linestyle='-.', axis='y')

# Meses de referência do gráfico
plt.xticks(ind, ('Oct 2018', 'Nov 2018', 'Dez 2018'), color='#292929', fontsize=18)

plt.yticks(np.arange(0, 101, 25), color='#292929', fontsize=18)
plt.legend((p1[0], p2[0]), ('% Atingidos', '% Violados'), prop={'size': 16}, loc=4)
plt.gca().yaxis.set_major_formatter(PercentFormatter())
for p in plt.gca().patches:
    if p.get_height() > 50:
        plt.gca().annotate(str(p.get_height())+'%', (p.get_x() + 0.77*p.get_width()/2, p.get_y() + p.get_height()*2/3), fontsize=18)

plt.show()
```
