```python
import matplotlib.pyplot as plt

colors = ['#99cc99', '#ff7f7f']

fig, axs = plt.subplots(1, 2)

# Porcentagem do primeiro bloco: [Positivo, Negativo]
sizes = [91, 9]
axs[0].pie(sizes, colors=colors,
          startangle=30, pctdistance=0.9, wedgeprops=dict(width=0.3, edgecolor=(0,0,0,0)))


# Porcentagem do segundo bloco: [Positivo, Negativo]
sizes = [70, 30]
axs[1].pie(sizes, colors=colors,
          startangle=30, pctdistance=0.9, wedgeprops=dict(width=0.3, edgecolor=(0,0,0,0)))

plt.suptitle('Índice de Satisfação', fontsize=16, color='#292929')

# Título do primeiro bloco
axs[0].set_title('Viagem e Turismo', color='#292929')

# Porcentagem descrita do primeiro bloco
axs[0].text(0, 0, '91%', ha='center', fontsize=14, color='#292929')

# Título do segundo bloco
axs[1].set_title('Empresa', color='#292929')

# Porcentagem descrita do segundo bloco
axs[1].text(0, 0, '70%', ha='center', fontsize=14, color='#292929')

plt.show()
```
