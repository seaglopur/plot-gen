```python
import matplotlib.pyplot as plt

colors = ['#99cc99', '#ff7f7f']

fig, axs = plt.subplots(1, 2)
sizes = [91, 9]
axs[0].pie(sizes, colors=colors,
          startangle=30, pctdistance=0.9, wedgeprops=dict(width=0.3, edgecolor=(0,0,0,0)))


sizes = [70, 30]
axs[1].pie(sizes, colors=colors,
          startangle=30, pctdistance=0.9, wedgeprops=dict(width=0.3, edgecolor=(0,0,0,0)))

plt.suptitle('Índice de Satisfação', fontsize=16, color='white')
axs[0].set_title('Viagem e Turismo', color='white')
axs[0].text(0, 0, '91%', ha='center', fontsize=14, color='white')
axs[1].set_title('Viajanet', color='white')
axs[1].text(0, 0, '70%', ha='center', fontsize=14, color='white')

plt.show()
```
